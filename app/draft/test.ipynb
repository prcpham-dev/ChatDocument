```python
import requests
```

```python
URL = "https://theforevercanvas.com/api"

def get_stream(query: str):
    s = requests.Session()
    session = "123"

    # res = s.post(f"{URL}/model_activation", json={"session_id": session})
    # assert res.status_code == 200

    with s.get(
        f"{URL}/chat_completion/",
        stream=True,
        params={"question": query, 'session_id': session},
    ) as r:
        for line in r.iter_content(decode_unicode=True, chunk_size=1):
            print(line, end="", flush=True)

get_stream("tell me a 100 words story")
```
